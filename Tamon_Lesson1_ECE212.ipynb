#### Exercise	3:	Arithmetic	(Solved,	22	Lines)

Create	a	program	that	reads	two integers,	a	and	b,	from	the	user.	Your	program	should
compute	and	display:

•The	sum	of	a	and	b

•The	difference	when	b	is	subtracted	from	a

•The	product	of	a	and	b

•The	quotient	when	a	is	divided	by	b

•The	remainder	when	a	is	divided	by	b

•The	result	of	log10	a

•The	result	of	ab

```python
# Two integers
a = float(input("The value of a is: "))
b = float(input("The value of b is: "))

sum = a + b
print ("The sum of a and b: ", sum)

difference = a - b
print ("The difference when b is subtracted from a: ", difference)

product = a * b
print ("The product of a and b: ", product)

quotient = a / b
print ("The quotient when a is divided by b: ", quotient)

remainder = a % b
print ("The remainder when a is divided by b: ", remainder)

import math
log = math.log10(a)
print ("The result of log10 a: ", log)

ab =  math.log10(product)
print ("The result of ab: ", ab)

```

#### Exercise	4:	Area	and	Volume	(15	Lines)


Write	a	program	that	begins	by	reading	a	radius,	r,	from	the	user.	The	program	will
continue	by	computing	and	displaying	the	area	of	a	circle	with	radius	r	and	the

calculations.

```python
from math import pi
radius = float(input('Radius of a circle: '))
area = pi * radius **2
volume = (4/3) * (pi * radius ** 3)
print("Area is: ", area)
print("Volume is: ", volume)
```
